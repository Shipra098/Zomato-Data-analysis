```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Indian-Resturants.csv')
```

```python
df
```

```python
# checking the data types of a data.
df.dtypes
```

```python
# checking the total duplicated values in a data.
df.duplicated().sum()
```

```python
# droping the duplicated values for data cleaning
df = df.drop_duplicates()
```

```python
df
```

```python
# finding the missing values.
df.isnull().sum()
```

```python
# filling the null values of a data.
df.zipcode = df.zipcode.fillna(df.zipcode.mode()[0])
```

```python
df.cuisines = df.cuisines.fillna(df.cuisines.mode()[0])
```

```python
df.timings = df.timings.fillna(df.timings.mode()[0])
```

```python
df.address =df.address.bfill()
```

```python
# droping the columns which is not in use.
df.drop('opentable_support',axis=1,inplace=True)
df.drop('url',axis=1,inplace=True)
```

```python
# overall average rating of Indian restraunt
df.aggregate_rating.mean()
```

```python
# average rating of restraunt in different cities.
df.groupby('city')['aggregate_rating'].mean().sort_values(ascending=False).head(10)
```

```python
df.groupby('city')['aggregate_rating'].mean().sort_values(ascending=False).head(10).plot(kind='pie',autopct='%1.1f%%')
```

```python
# Rating reviews of restraunts in diff cities.
df.groupby('rating_text')['city'].value_counts().reset_index().sort_values(by='city',ascending=False)
```

```python
df.columns
```

```python
# restraunts ratiings
df.groupby('name')['aggregate_rating'].mean().sort_values(ascending=False)
```

```python
# Barplot of Indian restraunts which has given top 10 highest votes.
df.groupby('name')['votes'].mean().sort_values(ascending=False).head(10).plot(kind='bar',color='r')
```

```python
# Identifying the most popluar cuisines .
df.cuisines.value_counts()
```

* North Indian is the most popular cuisines among all.

```python
df.price_range.mean()
```

```python
# Average cost for two persons according to its city and cuisines.
df.groupby(['city','cuisines'])['average_cost_for_two'].mean().sort_values(ascending=False).reset_index()
```

```python
# Top 10 name of restraunts who are in high demand.
sns.barplot(x=df.name.value_counts().head(10).index,y=df.name.value_counts().head(10))
plt.tick_params(rotation=50,axis='x')
plt.title('Top 10 Restaurant Chains Based on Number of Outlets')
```

* Domino's Pizza has most restraunts into diff cities.
